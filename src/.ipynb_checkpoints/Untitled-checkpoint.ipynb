```python
import pandas as pd
import numpy as np
paises = pd.read_csv("../paises.csv")
paises.head()
paises.columns = [c.lstrip() for c in paises.columns]
dict_paises ={}
def lower(x):
    return x.lower()
paises.name = paises.name.apply(lower)
paises.head()
i = 0
for ele in paises.name:
    dict_paises[ele] = paises.iso3.iloc[i]
    i += 1
dict_paises
```
